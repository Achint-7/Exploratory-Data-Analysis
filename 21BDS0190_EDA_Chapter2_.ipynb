#1. Upload the Dataset to Google Colab

```python
#Load Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from GitHub
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/lithium-ion%20batteries.csv'
df = pd.read_csv(url)
print('ACHINT TRIPATHI 21BDS0190')
```

```python
df.head()
```

#2. Identifing Missing Values

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
# Verify the replacement
print(df.head())
```

```python
# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)
```

#3. Transformation Techniques

#3.1 Data Deduplication

```python
# Remove duplicate rows
df_dedup = df.drop_duplicates()
df_dedup.head()
```

#3.2 Replacing Values

```python
# Replace specific values
df_replaced = df.replace({'C2/c': 'C2_c', 'P21': 'P_21', 'P21/c':'P21_c'})
df_replaced.head()
```

#3.3 Handling Missing Data

```python
# Fill missing values in 'Formula' with the mode
df['Formula'] = df['Formula'].fillna(df['Formula'].mode()[0])
```

```python
# Convert 'E Above Hull (eV)' to numeric, forcing errors to NaN
df['E Above Hull (eV)'] = pd.to_numeric(df['E Above Hull (eV)'], errors='coerce')

# Fill missing values with the mean
df['E Above Hull (eV)'] = df['E Above Hull (eV)'].fillna(df['E Above Hull (eV)'].mean())
```

```python
# Fill missing values in 'Band Gap (eV)' with the mean
df['Band Gap (eV)'] = df['Band Gap (eV)'].fillna(df['Band Gap (eV)'].mean())
```

```python
# Fill missing values in 'Crystal System' with the mode
df['Crystal System'] = df['Crystal System'].fillna(df['Crystal System'].mode()[0])
```

```python
# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)
```

#4. Discretization and Binning

```python
# Discretize 'Formation Energy (eV)' into bins
df['Energy_bin'] = pd.cut(df['Formation Energy (eV)'], bins=3, labels=['Low', 'Medium', 'High'])
```

```python
# Discretize 'Formation Energy (eV)' into quantile-based bins
df['Energy_quantile'] = pd.qcut(df['Formation Energy (eV)'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
```

#5. Outlier Detection and Filtering

```python
# Detect outliers using Z-score
df['z_score'] = (df['Formation Energy (eV)'] - df['Formation Energy (eV)'].mean()) / df['Formation Energy (eV)'].std()
outliers = df[np.abs(df['z_score']) > 3]

# Filter out outliers
df_filtered = df[np.abs(df['z_score']) <= 3]
```

```python
# Detect outliers using IQR
Q1 = df['Formation Energy (eV)'].quantile(0.25)
Q3 = df['Formation Energy (eV)'].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = df[(df['Formation Energy (eV)'] < (Q1 - 1.5 * IQR)) | (df['Formation Energy (eV)'] > (Q3 + 1.5 * IQR))]

# Filter out outliers using IQR
df_filtered_iqr = df[~((df['Formation Energy (eV)'] < (Q1 - 1.5 * IQR)) | (df['Formation Energy (eV)'] > (Q3 + 1.5 * IQR)))]
```

```python
# Plotting
plt.figure(figsize=(14, 6))

# Plot for Z-score method
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x=df.index, y='Formation Energy (eV)', label='Data')
sns.scatterplot(data=outliers, x=outliers.index, y='Formation Energy (eV)', color='red', label='Outliers')
plt.title('ACHINT TRIPATHI 21BDS0190 Outliers Detection using Z-score')
plt.xlabel('Index')
plt.ylabel('Formation Energy (eV)')
```

```python

# Plot for IQR method
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x=df.index, y='Formation Energy (eV)', label='Data')
sns.scatterplot(data=outliers_iqr, x=outliers_iqr.index, y='Formation Energy (eV)', color='red', label='Outliers')
plt.title('ACHINT TRIPATHI 21BDS0190 Outliers Detection using IQR')
plt.xlabel('Index')
plt.ylabel('Formation Energy (eV)')

plt.tight_layout()
plt.show()
```
