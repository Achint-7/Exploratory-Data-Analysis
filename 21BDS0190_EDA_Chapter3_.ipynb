```python
#Load Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from GitHub
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/lithium-ion%20batteries.csv'
df = pd.read_csv(url)
print('ACHINT TRIPATHI 21BDS0190')
```

```python
df.head()
```

#2. Identifing Missing Values

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
# Verify the replacement
print(df.head())
```

```python
# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)
print('ACHINT TRIPATHI 21BDS0190')
```

#3. Transformation Techniques

```python
# Remove duplicate rows
df_dedup = df.drop_duplicates()
df_dedup.head()
```

```python
# Replace specific values
df_replaced = df.replace({'C2/c': 'C2_c', 'P21': 'P_21', 'P21/c':'P21_c'})
df_replaced.head()
```

```python
# Fill missing values in 'Formula' with the mode
df['Formula'] = df['Formula'].fillna(df['Formula'].mode()[0])
print('ACHINT TRIPATHI 21BDS0190')
```

```python
# Convert 'E Above Hull (eV)' to numeric, forcing errors to NaN
df['E Above Hull (eV)'] = pd.to_numeric(df['E Above Hull (eV)'], errors='coerce')

# Fill missing values with the mean
df['E Above Hull (eV)'] = df['E Above Hull (eV)'].fillna(df['E Above Hull (eV)'].mean())
print('ACHINT TRIPATHI 21BDS0190')
```

```python
# Fill missing values in 'Band Gap (eV)' with the mean
df['Band Gap (eV)'] = df['Band Gap (eV)'].fillna(df['Band Gap (eV)'].mean())
print('ACHINT TRIPATHI 21BDS0190')
```

```python
# Fill missing values in 'Crystal System' with the mode
df['Crystal System'] = df['Crystal System'].fillna(df['Crystal System'].mode()[0])
print('ACHINT TRIPATHI 21BDS0190')
```

```python
# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)
print('ACHINT TRIPATHI 21BDS0190')
```

3.3 Correlation Analysis

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation = numeric_df.corr(method='pearson')
print("Correlation matrix:\n", correlation)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm', xticklabels=correlation.columns, yticklabels=correlation.columns)
plt.title("21BDS0190 ACHINT-TRIPATHI Correlation Heatmap")
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

#TIME SERIES ANALYSIS

```python
%matplotlib inline
```

Since our dataset doesn't have a date column, we'll create a dummy date coloumn with a time index for 50 time points.

```python
# Simulating a time series
dates = pd.date_range(start='2020-01-01', periods=50, freq='D')
data = {
    'Formation Energy (eV)': np.random.normal(loc=-2.7, scale=0.1, size=50),
    'E Above Hull (eV)': np.random.normal(loc=0.01, scale=0.005, size=50),
    'Band Gap (eV)': np.random.normal(loc=3.0, scale=0.5, size=50),
    'Density (gm/cc)': np.random.normal(loc=3.0, scale=0.1, size=50),
}
df_batteries = pd.DataFrame(data, index=dates)
df_batteries['Weekday Name'] = df_batteries.index.day_name()
df_batteries.head()
```

#Visualizing the Time Series Data

```python
plt.figure(figsize=(16, 8))
sns.lineplot(data=df_batteries['Formation Energy (eV)'])
plt.title('21BDS0190 Formation Energy Over Time')
plt.xlabel('Time Index')
plt.ylabel('Formation Energy (eV)')
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

#Cumulative Sum Plot

```python
cumulative_sum = df_batteries['Formation Energy (eV)'].cumsum()

plt.figure(figsize=(16, 8))
sns.lineplot(data=cumulative_sum)
plt.title('21BDS0190 Cumulative Sum of Formation Energy Over Time')
plt.xlabel('Time Index')
plt.ylabel('Cumulative Formation Energy (eV)')
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

```python
#Adding Year, Month, and Weekday Name
df_batteries['Year'] = df_batteries.index.year
df_batteries['Month'] = df_batteries.index.month
df_batteries['Weekday Name'] = df_batteries.index.day_name()
print('ACHINT TRIPATHI 21BDS0190')
```

Visualizing Different Columns

```python
#Visualizing Different Columns
cols_to_plot = ['Formation Energy (eV)', 'E Above Hull (eV)', 'Band Gap (eV)', 'Density (gm/cc)']
axes = df_batteries[cols_to_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(14, 6), subplots=True)

for ax in axes:
    ax.set_ylabel('Values')
plt.tight_layout()
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

Box Plots by Month

```python
sns.boxplot(data=df_batteries, x='Month', y='Formation Energy (eV)')
plt.title('21BDS0190 Box Plot of Formation Energy by Month')
plt.ylabel('Formation Energy (eV)')
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

```python
# Creating a categorical variable based on Band Gap
def categorize_band_gap(value):
    if value < 2:
        return 'Low Band Gap'
    elif value < 4:
        return 'Medium Band Gap'
    else:
        return 'High Band Gap'
print('ACHINT TRIPATHI 21BDS0190')
df_batteries['Band Gap Category'] = df_batteries['Band Gap (eV)'].apply(categorize_band_gap)
df_batteries.head()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 7), sharex=True)
for name, ax in zip(['Formation Energy (eV)', 'E Above Hull (eV)', 'Band Gap (eV)'], axes):
    sns.boxplot(data=df_batteries, x='Band Gap Category', y=name, ax=ax)
    ax.set_ylabel(name)
    ax.set_title(f'21BDS0190 {name} by Band Gap Category')
    if ax != axes[-1]:
        ax.set_xlabel('')
plt.tight_layout()
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```

```python
grouped_data = df_batteries.groupby('Weekday Name')['Formation Energy (eV)'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_batteries, x='Weekday Name', y='Formation Energy (eV)')
plt.title('21BDS0190 Box Plot of Formation Energy by Weekday')
plt.ylabel('Formation Energy (eV)')
plt.xlabel('Weekday Name')
plt.xticks(rotation=45)
plt.show()
print('ACHINT TRIPATHI 21BDS0190')
```
